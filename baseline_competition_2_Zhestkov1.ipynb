## Prepare data

Function for reading the data and saving it into a sparse matrix

```python
from scipy import sparse

def read_data_from_file(filename):
    values = list()
    rows = list()
    cols = list()
    
    header = True
    for line in open(filename):
        if header:
            header = False
            continue
        value, row, col = [int(x) for x in line.strip().split(',')]
        row -= 1
        col -= 1
        values.append(value)
        rows.append(row)
        cols.append(col)
        
    return sparse.csr_matrix((values, (rows, cols)), shape=(10000, 25640))
```

Function for reading labels

```python
def read_labels_from_file(filename):
    labels = zeros((10000, 83)).astype(int)

    header = True
    for line in open(filename):
        if header:
            header = False
            continue
        row, indeces = line.strip().split(',')
        row = int(row) - 1
        indeces = [int(x) - 1 for x in indeces.split()]
        labels[row, indeces] = 1
    
    return labels
```

## Try to find optimal coefficient for division train and test set.

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.cross_validation import train_test_split
import numpy as np
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import fbeta_score, make_scorer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import f1_score, make_scorer
from sklearn.naive_bayes import GaussianNB, MultinomialNB,BernoulliNB

class BaselineBayesClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, n_jobs=-1, alpha=0.1):
        self.n_jobs = n_jobs
        self.alpha = alpha
    
    def fit(self, X, y):
        self.clf = OneVsRestClassifier(
                MultinomialNB(alpha=self.alpha, class_prior=None, fit_prior=True),
                n_jobs=self.n_jobs,
        )
        self.clf.fit(X, y)
    
    def predict(self, X):
        return self.clf.predict(X)
```

```python
means = []
stds = []
from sklearn.grid_search import GridSearchCV
for k in arange (0.2,0.8,0.1):
    X_trainNEW, X_testNEW, y_trainNEW, y_testNEW =  train_test_split(X_train, y_train, test_size=k)
    print X_trainNEW.shape, y_trainNEW.shape, X_testNEW.shape, y_testNEW.shape

    grid_searcher = GridSearchCV(
            estimator=BaselineBayesClassifier(),
            param_grid={'alpha': [0.1]},
            scoring=make_scorer(f1_score, average='samples'),
            cv=2,
    )
    
    grid_searcher.fit(X_trainNEW, y_trainNEW)

    for elem in grid_searcher.grid_scores_:
        print elem
        means.append(mean(elem.cv_validation_scores))
        stds.append(sqrt(var(elem.cv_validation_scores, ddof=1)))
means = array(means)
stds = array(stds)
figure(figsize=(8, 6))
plot(means)
fill_between(arange(0.2,0.8,0.1), means + stds, means - stds, alpha = 0.1, facecolor='blue')
xlabel('k-division coefficient, share of test set')
ylabel('Accuracy')
```

## KNN baseline

Class for fitting parameter k in multilabel KNN

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier

class BaselineKnnClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, k=1, n_jobs=3):
        self.k = k
        self.n_jobs = n_jobs
    
    def fit(self, X, y):
        self.clf = OneVsRestClassifier(
                KNeighborsClassifier(n_neighbors=self.k, weights='distance'),
                n_jobs=self.n_jobs,
        )
        self.clf.fit(X, y)
    
    def predict(self, X):
        return self.clf.predict(X)
```

Find optimal neigbours number

```python
from sklearn.metrics import f1_score, make_scorer
from sklearn.grid_search import GridSearchCV

grid_searcher = GridSearchCV(
        estimator=BaselineKnnClassifier(),
        param_grid={'k': logspace(0, 3, 10).astype(int)},
        scoring=make_scorer(f1_score, average='samples'),
        cv=2,
)

grid_searcher.fit(X_train, y_train)
print grid_searcher.best_params_
print grid_searcher.best_score_
print grid_searcher.best_estimator_
```

Function for a nice visualization

```python
def plot_grid_searcher_results(grid_seacher, log_scale):
    means = list()
    param_values = list()

    for score in grid_searcher.grid_scores_:
        print score.cv_validation_scores
        means.append(score.mean_validation_score)
        param_values.append(score.parameters.values()[0])

    parameter_name = grid_searcher.grid_scores_[0].parameters.keys()[0]
    means = array(means)
    
    plot(param_values, means)
    if log_scale:
        xscale('log')
    xlabel(parameter_name)
    ylabel('Quality')
```

Visualization

```python
plot_grid_searcher_results(grid_searcher, log_scale=True)
```

We can see that low number of neighbours is optimal.

```python
grid_searcher = GridSearchCV(
        estimator=BaselineKnnClassifier(),
        param_grid={'k': range(1, 10)},
        scoring=make_scorer(f1_score, average='samples'),
        cv=2,
)

grid_searcher.fit(X_train, y_train)
plot_grid_searcher_results(grid_searcher, log_scale=False)
```

## SVM baseline

Class for fitting C parameter in multilabel SVM

```python
from sklearn.svm import LinearSVC

class BaselineLinearClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, C=1.0, n_jobs=-1):
        self.C = C
        self.n_jobs = n_jobs
    
    def fit(self, X, y):
        self.clf = OneVsRestClassifier(LinearSVC(C=self.C), self.n_jobs)
        self.clf.fit(X, y)
    
    def predict(self, X):
        return self.clf.predict(X)
```

Fit C parameter

```python
from sklearn.metrics import f1_score, make_scorer
from sklearn.grid_search import GridSearchCV

grid_searcher = GridSearchCV(
        estimator=BaselineLinearClassifier(),
        param_grid={'C': [0.0001, 0.001, 0.005, 0.01, 0.1, 1, 5, 10, 100, 1000,2000,10000]},
        scoring=make_scorer(f1_score, average='samples'),
        cv=5
)

grid_searcher.fit(X_train, y_train)
```

Visualization

```python
plot_grid_searcher_results(grid_searcher, log_scale=True)
```

We can see that optimal value $C>1000$. Train classifier and save the results

```python
clf = BaselineLinearClassifier(C=1000, n_jobs=3)
clf.fit(X_train, y_train)
y_test_svm = clf.predict(X_test)
write_labels_to_file(y_test_svm, 'D:\Skoltech\Study\Machine learning\competition 2\y_test_svm.csv')
```
